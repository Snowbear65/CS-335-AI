```python
# CS 335 — Introduction to Artificial Intelligence
# API Assignment Starter Code — Northeastern Illinois University
#
# TODO: Replace BASE_URL, endpoints, params, and payload fields
#       with values from your chosen API's documentation.
# ─────────────────────────────────────────────────────────────

import os
import json
import requests
from dotenv import load_dotenv
from IPython.display import Image, display

load_dotenv()

#API_KEY = os.getenv("MY_API_KEY") NO API KEY
#if not API_KEY:
#    raise ValueError("API key not found. Did you copy .env.example to .env?")

# TODO: Replace with your API's base URL
BASE_URL = "https://pokeapi.co/api/v2"

# update or extend per call if your API requires it
HEADERS = {
    "Content-Type": "application/json",
}


def divider(label):
    print(f"\n{'=' * 50}\n{label}\n{'=' * 50}")


# ── Call 1: GET request ───────────────────────────────────
# Use for retrieving data without a request body.
# TODO: Update url and params.
def call_one_get():
    divider("CALL 1 — GET Request")

    url = f"{BASE_URL}/pokemon/bulbasaur" #Bulbasaur No.1 Pokemon Baby!
    params = {}  # TODO: update these

    response = requests.get(url, headers=HEADERS, params=params)

    if response.status_code == 200:
      data = response.json()
      result = {
            "name": data["name"],
            "types": [t["type"]["name"] for t in data["types"]],
            "abilities": [a["ability"]["name"] for a in data["abilities"]],
            "front_sprite": data["sprites"]["front_default"]
        }
      print(json.dumps(result, indent=2, ensure_ascii=False))
      if data["sprites"]["front_default"]:
            display(Image(url=data["sprites"]["front_default"], width=288))
    else:
        print(f"[ERROR] {response.status_code}: {response.text}")


# ── Call 2: POST request ──────────────────────────────────
# Use for sending data to the API (e.g., a prompt or input text).
# TODO: Update url and payload fields.
def call_two_post():
    divider("CALL 2 — POST Request")

    url = f"{BASE_URL}/ability/overgrow"
    response = requests.get(url, headers=HEADERS)

    if response.status_code == 200:
        data = response.json()

        english_effect = "No English effect found."
        for entry in data["effect_entries"]:
            if entry["language"]["name"] == "en":
                english_effect = entry["effect"]
                break

        result = {
            "ability": data["name"],
            "effect": english_effect
        }

        print(json.dumps(result,  indent=2, ensure_ascii=False))  # TODO: update key
    elif response.status_code == 401:
        print("[ERROR] 401 Unauthorized — check your API key in .env")
    elif response.status_code == 429:
        print("[ERROR] 429 Rate Limited — wait and retry")
    else:
        print(f"[ERROR] {response.status_code}: {response.text}")


# ── Call 3: Parameterized POST ────────────────────────────
# Same as Call 2 but accepts dynamic input to show varied output.
# TODO: Update url and payload fields.
def call_three_parameterized():
    divider("CALL 3 — Parameterized")

    try:
        user_input = int(input("Enter a Pokémon # (1–1024): "))

        if user_input < 1 or user_input > 1024:
            print("[ERROR] Please enter a number between 1 and 1024.")
            return

        url = f"{BASE_URL}/pokemon/{user_input}"
        response = requests.get(url, headers=HEADERS)

        if response.status_code == 200:
            data = response.json()
            result = {
                "name": data["name"],
                "types": [t["type"]["name"] for t in data["types"]],
                "abilities": [a["ability"]["name"] for a in data["abilities"]],
                "front_sprite": data["sprites"]["front_default"]
            }
            print(json.dumps(result, indent=2, ensure_ascii=False))

            if data["sprites"]["front_default"]:
                display(Image(url=data["sprites"]["front_default"], width=288))

        elif response.status_code == 404:
            print("[ERROR] Pokémon not found.")
        elif response.status_code == 429:
            print("[ERROR] 429 Rate Limited — slow down and retry")
        else:
            print(f"[ERROR] {response.status_code}: {response.text}")

    except ValueError:
        print("[ERROR] Invalid input. Please enter a number.")


if __name__ == "__main__":
    call_one_get()
    call_two_post()
    call_three_parameterized()
```
